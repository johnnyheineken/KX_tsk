```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('dataset.csv', sep =';', header=1 )
```

```python
data.shape
```

```python
data.head(100)
```

```python
labels = ['Green frogs','Brown frogs','Common toad','Fire-bellied toad','Tree frog','Common newt','Great crested newt']
```

```python
for label in labels:
    print(label)
    print(data[label].mean())
    data[label].value_counts().plot(kind = 'bar')
    print('---------------------------------------------')
```

```python
corr = data[labels].corr(method ='pearson')
corr=round(corr,2)
#corr.style.background_gradient(cmap='coolwarm')
sns.set(rc={'figure.figsize':(15,15)})
sns.heatmap(corr, annot=True, xticklabels=corr.columns, yticklabels=corr.columns)
```
